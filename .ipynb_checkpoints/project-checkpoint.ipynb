```python
from konlpy.tag import Twitter
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from konlpy.tag import Okt

from sklearn.feature_extraction.text import CountVectorizer
```

```python
def get_noun(text):
    tokenizer = Twitter()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

```python
def get_noun2(text):
    tokenizer = Hannanum()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

```python
def get_noun3(text):
    tokenizer = Okt()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

```python
def get_morphs3(text):
    tokenizer = Okt()
    morphs = tokenizer.morphs(text)
    return [n for n in morphs]
```

```python
def get_noun4(text):
    tokenizer = Komoran()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
```

```python
text_clf_svm = Pipeline([('vect', CountVectorizer(tokenizer=get_noun3)),
                        ('tfidf', TfidfTransformer()),
                        ('clf-svm', SGDClassifier(loss='hinge',
                                                 alpha=1e-3, random_state=42,
                                                 max_iter = 1500))])
text_clf_svm = text_clf_svm.fit(train_set["news"], train_set['label'])
```

```python
predicted_svm = text_clf_svm.predict(test_set['news'])
np.mean(predicted_svm == test_set['label'])
```
